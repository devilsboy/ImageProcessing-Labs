# **QUESTION 1**

**USING PIL**

```python
from PIL import Image
import requests
from io import BytesIO

url = requests.get('https://i.imgur.com/YZwMI6f.jpg')
img = Image.open(BytesIO(url.content))
img
```

**WIDTH AND SIZE**

```python
w,h = img.size
print("Width:",w)
print("Hieght:",h)
```

**SAVING IMAGE**

```python
img.save("PIL.jpg")
```

**USING SKIMAGE**

```python
import skimage.io
import skimage.viewer
img = skimage.io.imread(fname="PIL.jpg")
img
```

```python
skimage.io.imshow(img) 
skimage.io.show()
```

**SAVING IMAGE**

```python
skimage.io.imsave("skimage.jpg",arr = img)
```

# **QUESTION 2**

```python
import numpy as np
rows = 512
cols = 512
image = open('barbara_gray.raw')
print(image)
```

```python
img = np.fromfile(image, dtype = np.uint8, count = rows * cols)
```

```python
print("1D Array Dimensions: ",img.ndim)
img.size
```

**CONVERTING 1D ARRAY TO 2D ARRAY**

```python
img.shape = (img.size//cols, cols)
print("2D Array Dimensions: ",img.ndim)
```

**SAVING RAW IMAGE**

```python
img.astype('int8').tofile('NewImage.raw')
image.close
```
